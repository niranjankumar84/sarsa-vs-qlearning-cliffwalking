```python
import gymnasium as gym
import random
import numpy as np
```

```python
#create object of environment
env=gym.make("CliffWalking-v1")
print(env.observation_space.n) #no-of-states
print(env.action_space.n)#no-of-actions
starting_state,_=env.reset() #starting single state values(3,0=>12*3+0=36)
print(starting_state)
print(_)
```

```python
#step1
# Q-table =>stores Q-values
Q=np.zeros((48,4))
```

# 1.SARSA ALGORITHMS

```python
gamma=0.99
alpha=0.1
episodes=500
epsilon=0.1
```

```python
# define the policy:-epsilon greedy policy:-state->action

def epsilon_greedy(state):
    if random.random()<epsilon:
        return env.action_space.sample() #it generate random action=>explore
    else:
        return np.argmax(Q[state]) #exploit
    
        
```

```python
for episode in range(episodes):
    render=(episode%50==0)
    if render:
        env=gym.make("CliffWalking-v1",render_mode="human")
    else:
        env=gym.make("CliffWalking-v1")
    done=False
    state,_=env.reset() # return starting index
    action=epsilon_greedy(state)
    total_reward=0
    episode_len=0
    
    while not done:
        next_state,reward,terminated,truncated,_=env.step(action)
        done=terminated or truncated
        next_action=epsilon_greedy(next_state)
    
        #SARSA updation
        Q[state,action]+=alpha*(reward+gamma*(Q[next_state,next_action]-Q[state,action]))
        state=next_state
        action=next_action
    
        total_reward+=reward
        episode_len+=1
    print(f"episode={episode+1}/500 ,total_reward={total_reward} & episode_length={episode_len}")
    env.close()
        
    
```

```python
# what did our agenet learn
env=gym.make("CliffWalking-v1",render_mode="human")
state,_=env.reset()
done=False
total_reward=0
episode_len=0

while not done:
    action=np.argmax(Q[state])
    state,reward,terminated,truncate,_=env.step(action)
    done=terminated or truncate

    total_reward+=reward
    episode_len+=1
print(f"total_rewards={total_reward} & episode_len={episode_len}")
env.close()
```
